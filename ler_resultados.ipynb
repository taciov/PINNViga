```python
import pandas as pd
import numpy as np

## OBTER DIFERENÃ‡AS

def obter_diff(df_diff):
    u_max = float(df_diff['u_max'].values[-1])
    u_min = float(df_diff['u_min'].values[-1])
    theta_max = float(df_diff['theta_max'].values[-1])
    theta_min = float(df_diff['theta_min'].values[-1])
    V_max = float(df_diff['V_max'].values[-1])
    V_min = float(df_diff['V_min'].values[-1])
    mf_max = float(df_diff['mf_max'].values[-1])
    mf_min = float(df_diff['mf_min'].values[-1])
    tempo = float(df_diff['tempo'].values[-1])

    df_diff['u_max'] = round(100 * abs(df_diff['u_max'] - u_max) / abs(u_max), 2)
    df_diff['u_min'] = round(100 * abs(df_diff['u_min'] - u_min) / u_min, 2)
    df_diff['theta_max'] = round(100 * abs(df_diff['theta_max'] - theta_max) / theta_max, 2)
    df_diff['theta_min'] = round(100 * abs(df_diff['theta_min'] - theta_min) / theta_min, 2)
    df_diff['V_max'] = round(100 * abs(df_diff['V_max'] - V_max) / V_max, 2)
    df_diff['V_min'] = round(100 * abs(df_diff['V_min'] - V_min) / V_min, 2)
    df_diff['mf_max'] = round(100 * abs(df_diff['mf_max'] - mf_max) / mf_max, 2)
    df_diff['mf_min'] = round(100 * abs(df_diff['mf_min'] - mf_min) / mf_min, 2)
    df_diff['tempo'] = round(df_diff['tempo'] / tempo)

    return df_diff
```

```python
df_1_lbfgs = pd.read_json("./lbfgs_1.json")
df_1_lbfgs
```

```python
df_1_lbfgs_diff = obter_diff(df_1_lbfgs)
df_1_lbfgs_diff
```
